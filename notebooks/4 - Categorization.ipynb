# Сумма нормированных параметров
кво фото, кво отзывов, рейтинг

```python
places_df = pd.read_csv('../data/places_GP.csv')
```

```python
places_df.describe()
# .plot(kind='box', ylim = [0, 200])
```

```python
places_df_scale = places_df.copy()
```

```python
places_df_scale['count_review'] = places_df_scale['count_review'].apply(lambda x: x if x <=23 else 23)
places_df_scale['count_photo'] = places_df_scale['count_photo'].apply(lambda x: x if x <=27 else 27)
```

```python
places_df['rating_by_category'] = 0
from sklearn.preprocessing import MinMaxScaler, RobustScaler
scaler = MinMaxScaler()
places_df['normalized_rating'] = scaler.fit_transform(places_df[['count_review', 'rating', 'count_photo']].fillna(0)).sum(axis = 1)
for i in places_df.category.value_counts().keys():
    places_df.loc[places_df.category==i, 'rating_by_category'] = scaler.fit_transform(places_df.loc[places_df.category==i][['count_review', 'rating', 'count_photo']].fillna(0)).sum(axis = 1)
```

```python
places_df.sort_values('rating_by_category').loc[places_df.rating_by_category==3]
```
